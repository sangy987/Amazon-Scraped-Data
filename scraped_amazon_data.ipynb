```python
# -*- coding: utf-8 -*-
"""test.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1TTH1GE43CcaKtcKN2SAh1wb020SMrQw-
"""

import requests
from bs4 import BeautifulSoup
import re
s = input()
N = int(input())
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'})
results = []
for page in range(N):
  URL = "https://www.amazon.in/s?k=" +s + "&page="+str(page)
  r = requests.get(URL,headers=HEADERS)
  soup = BeautifulSoup(r.content, "html.parser")
  # print(soup)
  content = soup.find_all("div", attrs={'data-component-type':'s-search-result'})
  # print(content)
  for tag in content:
    
    # print(tag)    
    rating = tag.find("span", attrs={'class':'a-icon-alt'})
    if rating:
      rating = rating.text
    else:
      rating = None    
    number_of_reviewers = tag.find("span", attrs={'class':'a-size-base puis-light-weight-text s-link-centralized-style'})
    number_of_reviewers_alternate = tag.find("span", attrs={'class':'a-size-base s-underline-text'})
    if number_of_reviewers:
      number_of_reviewers = number_of_reviewers.text
    elif number_of_reviewers_alternate:
      number_of_reviewers = number_of_reviewers_alternate.text
    else:
      number_of_reviewers = None
    product = tag.find("span", attrs={'class':'a-size-base-plus a-color-base a-text-normal'})
    if product:
      product = product.text
    else:
      product = None
    
    price = tag.find("span", attrs={'class':'a-offscreen'})
    if price:
      price = price.text
    else:
      price = None
    results.append([product,rating,number_of_reviewers,price])
# print(results)
```

```python





import pandas as pd
df = pd.DataFrame (results,columns =['Product', 'Rating', 'Number Of Reviewers','Price'])
#if rating and number_of_reviewers is null then product is sponsored
df.head()
# df.shape
```
